```python
%matplotlib widget

import pandas as pd
import matplotlib.pyplot as plt
import mplcursors
```

```python
waterleveldf=pd.read_csv('Dataset/water_level_sept10_to_dec25_2024.csv')
waterleveldf.head()
```

Validating Data type

```python
print(waterleveldf.dtypes)
```

Converting to relevant data types:

```python
waterleveldf['dateTime']=pd.to_datetime(waterleveldf['dateTime'])
waterleveldf['value']=pd.to_numeric(waterleveldf['value'],errors='coerce')
waterleveldf.dtypes
```

Setting DateTime As Index

```python
waterleveldf.set_index('dateTime',inplace=True)
waterleveldf.head()
```

Water Level Subset

```python
waterlevel_dec10_dec12_2024= waterleveldf['2024-12-10':'2024-12-12']
```

```python
plt.style.use('ggplot')
fig,ax=plt.subplots()
waterlevel_dec10_dec12_2024.plot(ax=ax,linewidth=0.7)

# # Enable hover functionality to view values at each point and display them
# mplcursors.cursor(ax, hover=True)
# # cursor.connect("add", lambda sel: sel.annotation.set_text(f"Date: {sel.target[0]}\nValue: {sel.target[1]:.2f}"))
plt.show()
```
